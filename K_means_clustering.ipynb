```python
# prompt: create a synthetic pandas data frame with two numerical cols with 6 distinct clusters and do scatter plot
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic data with 4 clusters
X, y = make_blobs(n_samples=500, centers=6, cluster_std=0.40, random_state=0)

# Create a Pandas DataFrame
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
df['Cluster'] = y

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Feature1'], df['Feature2'], c=df['Cluster'], cmap='viridis')
plt.title('Synthetic Data with 4 Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
km = KMeans(n_clusters=6)
y_predicted = km.fit_predict(df[['Feature1','Feature2']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
# prompt: print unique values of df['cluster']

print(df['cluster'].unique())
```

```python
km.cluster_centers_
```

```python
plt.plot(km.cluster_centers_)
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]
df6 = df[df.cluster==5]
plt.scatter(df1['Feature1'],df1['Feature2'],color='green')
plt.scatter(df2['Feature1'],df2['Feature2'],color='red')
plt.scatter(df3['Feature1'],df3['Feature2'],color='black')
plt.scatter(df4['Feature1'],df4['Feature2'],color='yellow')
plt.scatter(df5['Feature1'],df5['Feature2'],color='blue')
plt.scatter(df6['Feature1'],df6['Feature2'],color='orange')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('feature1')
plt.ylabel('feature2')
plt.legend()
```

**Preprocessing using min max scaler**

#Elbow Plot
The technique to determine K, the number of clusters, is called the elbow method.``

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Feature1','Feature2']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
